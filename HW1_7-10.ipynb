# The Perceptron Learning Algorithm
In this problem, you will create your own target function f and data set D to see how the Perceptron Learning Algorithm works. Take d = 2 so you can visualize the problem, and assume X = [−1, 1] × [−1, 1] with uniform probability of picking each x ∈ X. 
<br>

In each run, choose a random line in the plane as your target function f (do this by taking two random, uniformly distributed points in [−1, 1] × [−1, 1] and taking the line passing through them), where one side of the line maps to +1 and the other maps to −1. Choose the inputs xn of the data set as random points (uniformly in X ), and evaluate the target function on each xn to get the corresponding output yn. 
<br>

Now, in each run, use the Perceptron Learning Algorithm to find g. Start the PLA with the weight vector w being all zeros (consider sign(0) = 0, so all points are initially misclassified), and at each iteration have the algorithm choose a point randomly
from the set of misclassified points. We are interested in two quantities: the number of iterations that PLA takes to converge to g, and the disagreement between f and g which is P[f(x) 6= g(x)] (the probability that f and g will disagree on their classification of a random point). You can either calculate this probability exactly, or
approximate it by generating a sufficiently large, separate set of points to estimate it.
<br>

In order to get a reliable estimate for these two quantities, you should repeat the experiment for 1000 runs (each run as specified above) and take the average over these runs.

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
d = 2
```

```python
x_lower, x_upper = -1, 1
y_lower, y_upper = -1, 1
```

```python
class Point:
    def __init__(self, x_coord, y_coord, label=0):
        self.x = x_coord
        self.y = y_coord
        self.label = label
        
    def __str__(self):
        return "({}; {})".format(self.x, self.y)
    
    def __repr__(self):
        return "({}; {}), label {}".format(self.x, self.y, self.label)
    
    def plot(self):
        color = 'green' if self.label == 1 else 'cyan'
        
        plt.plot([self.x], [self.y], marker='o', markersize=5, color=color)

        
        
```

```python
class Line:
    def __init__(self, point_1=None, point_2=None, intercept=None, slope=None):
        if point_1 and point_2:
            if point_1.x == point_2.x:
                self.slope = 0
            else:
                self.slope = (point_1.y - point_2.y) / (point_1.x - point_2.x)

            self.intercept = point_1.y - self.slope * point_1.x
        else:
            self.intercept = intercept
            self.slope = slope
            
        self.point_1 = point_1
        self.point_2 = point_2
    
    def f(self, x_array):
        return x_array * self.slope + self.intercept
        
    def plot_line(self, color='b'):
        x_s = np.arange(x_lower, x_upper, 0.1)
        y_s = self.f(x_s)
        plt.plot(x_s, y_s)
        
        if self.point_1 and self.point_2:
            plt.plot([self.point_1.x], [self.point_1.y], marker='o', markersize=5, color="red")
            plt.plot([self.point_2.x], [self.point_2.y], marker='o', markersize=5, color="red")
```

```python
np.random.seed(seed=12)


x_1, x_2 = np.random.uniform(low=x_lower,
                             high=x_upper,
                             size=2)

y_1, y_2 = np.random.uniform(low=y_lower,
                             high=y_upper,
                             size=2)

point_1 = Point(x_1, y_1)
point_2 = Point(x_2, y_2)
print('Point 1 is:', point_1)
print('Point 2 is:', point_2)
```

```python
f_line = Line(point_1=point_1, point_2=point_2)
```

```python
f_line.plot_line()
```

```python
class ProjectManager():
    
    def __init__(self, x_lower, x_upper, y_lower, y_upper):
        self.x_lower = x_lower
        self.x_upper = x_upper
        self.y_lower = y_lower
        self.y_upper = y_upper
        self.points = []
        self.w = [0, 0, 0] 
        
        x_1, x_2 = np.random.uniform(low=x_lower,
                             high=x_upper,
                             size=2)

        y_1, y_2 = np.random.uniform(low=y_lower,
                             high=y_upper,
                             size=2)
        self.f_line = Line(point_1=point_1, point_2=point_2)
        
    def generate_point(self, line):
        x = np.random.uniform(low=self.x_lower, high=self.x_upper, size=1)
        y = np.random.uniform(low=self.y_lower, high=self.y_upper, size=1)
        
        label = self.predict_label(Point(x, y), line)
        
        return Point(x, y, label)
        
    def generate_points(self, N):
        points = []
        for _ in range(N):
            a_point = self.generate_point(self.f_line)
            points.append(a_point)
        return points
    
    def predict_label(self, point, line):
        return np.sign(1*line.intercept + point.x*line.slope - 1*point.y)
    
    def update_w(self, misclassified_point):
        return np.add(self.w, [1, misclassified_point.x, misclassified_point.y]*misclassified_point.label)
    
    def PLA_iteration(self, iter_prop, plot):
        success = False
        if iter_prop > 0:
            g_intercept, g_slope = -self.w[0]/self.w[2], -self.w[1]/self.w[2]
            self.g_line = Line(intercept=g_intercept, slope=g_slope)
            
            if plot:
                self.g_line.plot_line(color='magenta')

            labels, real_labels = [], []
            for point in self.points:
                real_labels.append(point.label)
                labels.append(self.predict_label(point, self.g_line))
            misclassified_indexes = np.where(~np.equal(labels, real_labels))[0]
            
            if len(misclassified_indexes) > 0:
                to_fix_point = self.points[np.random.choice(misclassified_indexes)]
                self.w = self.update_w(to_fix_point)
            else:
                success = True
        
        else:
            to_fix_point = np.random.choice(self.points)
            self.w = self.update_w(to_fix_point)

        return success
    
    def run_program(self, N_points, plot=False):
        success = False
        iteration = 0
        self.points = self.generate_points(N_points)

        while not success:
            success = self.PLA_iteration(iteration/20, plot)
            if plot:
                if iteration > 0:
                    print("Iteration {}:".format(str(iteration)))
                self.f_line.plot_line(color='blue')
                for point in self.points:
                    point.plot()
                plt.xlim(self.x_lower, self.x_upper)
                plt.ylim(self.y_lower, self.y_upper)
                plt.show()
            
            iteration += 1
            
#             if iteration > 100:
#                 print(iteration)
            
        if success: 
            NUM_POINTS = N_points*100
            new_points = self.generate_points(N=NUM_POINTS)
            actual_labels, pred_labels = [], []
            for point in new_points:
                actual_labels.append(point.label)
                pred_labels.append(self.predict_label(point, self.g_line))
                
            misclassified_indexes = np.where(~np.equal(actual_labels, pred_labels))[0]

            prob = len(misclassified_indexes)/NUM_POINTS
            
        return iteration, prob
```

```python
pm = ProjectManager(x_lower, x_upper,
                   y_lower, y_upper)
```

## 7. Take N = 10. How many iterations does it take on average for the PLA to converge for N = 10 training points? 
<br>
## 8. Which of the following is closest to P[f(x) != g(x)] for N = 10?

```python
# one example with plots with 10 points
pm = ProjectManager(x_lower, x_upper,
                   y_lower, y_upper)
    
iteration_number, prob = pm.run_program(N_points = 10, plot=True)
    
```

```python
RUNS = 1000

number_of_iterations, probs = [], []

for run in tqdm(range(RUNS)):
    pm = ProjectManager(x_lower, x_upper,
                   y_lower, y_upper)
    
    iteration_number, prob = pm.run_program(N_points = 10, plot=False)
    number_of_iterations.append(iteration_number)
    probs.append(prob)
    
#     print("RUN: {}, iterations to converge: {}, probability of misclassified new points: {}".format(run, iteration_number, prob))
```

```python
np.mean(number_of_iterations) #-> b.10
```

```python
np.mean(probs)  #-> c. 0.1
```
